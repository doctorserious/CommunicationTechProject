```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

data = pd.read_csv('mittausdata.csv', delimiter=';')


x_train = data[['X', 'Y', 'Z']].values
y_train = data['Suunta'].values

y_train = y_train - 1

# Muunna luokat one-hot encoding -muotoon
y_train = tf.keras.utils.to_categorical(y_train, num_classes=6)

# Neuroverkon määrittely
model = Sequential([
    Dense(16, activation='relu', input_shape=(3,)),  # Piilokerros
    Dense(6, activation='softmax')                  # Ulostulokerros
])

model.compile(optimizer=Adam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])

# Mallin opetus
model.fit(x_train, y_train, epochs=40, batch_size=32)

# Mallin painojen ja bias-arvojen tallennus
weights = model.get_weights()
np.savez_compressed('weights.npz',
                    w_hidden=weights[0],
                    b_hidden=weights[1],
                    w_output=weights[2],
                    b_output=weights[3])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Lataa painot
loaded_weights = np.load('weights.npz')
w_hidden = loaded_weights['w_hidden']
b_hidden = loaded_weights['b_hidden']
w_output = loaded_weights['w_output']
b_output = loaded_weights['b_output']

def oma_relu(x):
    return max(0, x)

def oma_softmax(x):
    expo = [np.exp(i - np.max(x)) for i in x]  # Eksponentit, numeerisen vakauden vuoksi
    sum_expo = sum(expo)                       # Summataan eksponentit
    return [i / sum_expo for i in expo]        # Arvojen normalisointi todennäköisyyksiksi

def oma_aktivointi(testidata):
    # Piilokerros Dense 1 (ReLu)
    hidden_layer = []
    for i in range(len(b_hidden)):  # Iteroidaan piilokerroksen neuronit
        w_summa = b_hidden[i]
        for j in range(len(testidata)):
            w_summa += testidata[j] * w_hidden[j][i]
        hidden_layer.append(oma_relu(w_summa))

    # Ulostulokerros Dense 2 (Softmax)
    output_layer = []
    for i in range(len(b_output)):
        w_summa = b_output[i]
        for j in range(len(hidden_layer)):
            w_summa += hidden_layer[j] * w_output[j][i]
        output_layer.append(w_summa)

    # Softmax-aktivaatio ulostulokerrokseen
    output_layer = oma_softmax(output_layer)
    return output_layer

# Testidata ja tulokset
testidata = np.zeros((6, 3))
oma_tulos = np.zeros((6, 6))
malli_ennuste = np.zeros((6, 6))

# Täytä testidata
# Oletetaan, että `data` on pandas DataFrame, joka sisältää sarakkeet ['X', 'Y', 'Z']
testidata[0] = data[['X', 'Y', 'Z']].iloc[100].values
testidata[1] = data[['X', 'Y', 'Z']].iloc[300].values
testidata[2] = data[['X', 'Y', 'Z']].iloc[500].values
testidata[3] = data[['X', 'Y', 'Z']].iloc[400].values
testidata[4] = data[['X', 'Y', 'Z']].iloc[0].values
testidata[5] = data[['X', 'Y', 'Z']].iloc[200].values

# Tulokset ja visualisointi
for i in range(6):
    oma_tulos[i] = oma_aktivointi(testidata[i])
    malli_ennuste[i] = model.predict(testidata[i].reshape(1, -1))

    # Piirretään pylväsdiagrammi logaritmisella asteikolla
    plt.figure(figsize=(10, 5))
    indices = range(len(oma_tulos[i]))

    plt.bar(indices, oma_tulos[i], width=0.4, label='Oma Tulos', align='center')
    plt.bar([x + 0.4 for x in indices], malli_ennuste[i], width=0.4, label='Malli Ennuste', align='center')

    plt.yscale('log')  # Logaritminen asteikko
    plt.title(f'Tulosvertailu syötteelle {i+1}')
    plt.xlabel('Luokat')
    plt.ylabel('Todennäköisyys (logaritminen)')
    plt.xticks(indices)
    plt.legend()
    plt.show()

# Kokonaiserojen tulostaminen
ero = np.abs(oma_tulos - malli_ennuste)
print("\nKokonaiserot:\n", ero)
```

```python
# Painot tallennetusta tiedostosta
loaded_weights = np.load('weights.npz')
w_hidden = loaded_weights['w_hidden']
b_hidden = loaded_weights['b_hidden']
w_output = loaded_weights['w_output']
b_output = loaded_weights['b_output']

# Headerin luonti
with open("neuroverkonKertoimet.h", "w") as f:
    f.write("#ifndef NEUROVERKONKERTOIMET_H\n")
    f.write("#define NEUROVERKONKERTOIMET_H\n\n")

    f.write("// Piilokerroksen painot ja bias-arvot\n")
    f.write("float w_hidden[{}][{}] = {};\n".format(
        w_hidden.shape[0], w_hidden.shape[1],
        str(w_hidden.tolist()).replace('[', '{').replace(']', '}')
    ))
    f.write("float b_hidden[{}] = {};\n\n".format(
        b_hidden.shape[0],
        str(b_hidden.tolist()).replace('[', '{').replace(']', '}')
    ))

    f.write("// Ulostulokerroksen painot ja bias-arvot\n")
    f.write("float w_output[{}][{}] = {};\n".format(
        w_output.shape[0], w_output.shape[1],
        str(w_output.tolist()).replace('[', '{').replace(']', '}')
    ))
    f.write("float b_output[{}] = {};\n\n".format(
        b_output.shape[0],
        str(b_output.tolist()).replace('[', '{').replace(']', '}')
    ))

    f.write("#endif // NEUROVERKONKERTOIMET_H\n")
```
